```python
import pandas as pd
import numpy as np
d = pd.read_csv('CompanyJ_Twitterposts.csv')
d.head()
```

```python
d.drop(['Followers','Total Actions % Followers','Created Time', 'Public Clicks'] , axis =1, inplace= True )
d.head()
```

```python
df_null = pd.DataFrame({'Count': d.isnull().sum(), 'Percent': 100*d.isnull().sum()/len(d)})

df_null[df_null['Count'] > 0].sort_values(by=['Percent'], ascending=False)
```

```python
import re
message=[]
for msg in d.Message:
    x = re.sub(r'http\S+', '', msg)
    message.append(x)
d['Message'] = message
d.head()
```

```python
message=[]
for msg in d.Message:
    x = re.sub(r'\W+', ' ', msg)
    message.append(x)
d['Message'] = message
d.head()
```

```python
tweetcount = []
for msg in d.Message:
    x = len(msg)
    tweetcount.append(x)
d['Character Count'] = tweetcount
d.head()
```
