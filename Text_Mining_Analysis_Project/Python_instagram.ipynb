```python
import pandas as pd
import numpy as np
d = pd.read_csv('CompanyJ_Instaposts.csv')
d.head()
```

```python
d.drop(['Followers','Total Actions % Followers','Created Time', 'Media ID'] , axis =1, inplace= True )
d.head()
```

```python
df_null = pd.DataFrame({'Count': d.isnull().sum(), 'Percent': 100*d.isnull().sum()/len(d)})

df_null[df_null['Count'] > 0].sort_values(by=['Percent'], ascending=False)
```

```python
d.dropna(axis=0, inplace=True)
```

```python
tweetcount = []
for msg in d.Caption:
    x = len(msg)
    tweetcount.append(x)
d['Character Count'] = tweetcount
d.head()
```
